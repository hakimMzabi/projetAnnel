## Linear Simple 2D

### High gamma

```python
XTrain = np.array([
      [1],
      [2]
])
Y = np.array([
      [2],
      [3]
])

max = XTrain.max()
min = XTrain.min()
X = ((XTrain - min) / (max - min))

inputCount = 2
sampleCount = 1
gamma = 100

rbfModel = RBF.naive_rbf_train(X.flatten(),Y.flatten(),inputCount,sampleCount,gamma)

respredict = []
SampleVec = []
XToPredict = [[i / 100.0] for i in range(0,100) ]
XToPredictVec = []
count = 0
for r in XToPredict:
    SampleVec += r
    #CXToPredict = (c_double * len(SampleVec))(*SampleVec)
    res = RBF.naive_rbf_regression_predict(rbfModel, r)
    respredict.append(res)
    SampleVec = []
    
max = XTrain.max()
min = XTrain.min()
X = ((XTrain - min) / (max - min))

plt.scatter(X,Y,color="blue")
plt.scatter(XToPredict, respredict,color="lightblue")
plt.show()
plt.clf()
```

### Linear Simple 3D 

```python
X = np.array([ [1, 1], [2, 2], [3, 3] ]) 
Y = np.array([ 1, 2, 3 ])

XF = X.flatten()
YF = Y.flatten()

inputCountPerSample = 3
sampleCount = 2
gamma = 100

rbfModel = RBF.naive_rbf_train(XF,YF,inputCountPerSample,sampleCount,gamma)

fig = plt.figure()
ax = Axes3D(fig)
ax.scatter(X[:,0], X[:,1], Y, color="red")

XX, YY, ZZ = [], [], []

for i in range(0, 1000, 10):
    for j in range(0, 1000, 10):
        i /= 100
        j /= 100
        XX.append(i)
        YY.append(j)
        zz = RBF.naive_rbf_regression_predict(rbfModel, [i, j])
        ZZ.append(zz)

ax.plot_trisurf(XX, YY, ZZ, lw=0, color="grey", alpha=0.5)
plt.show()
```
