# Modèles Linéaires - Regression

```python
import CLibrary as CLib
import matplotlib.lines as mlines
import matplotlib.pyplot as plt
import numpy as np
import math
from mpl_toolkits.mplot3d import Axes3D
```

```python
def get(i, l):
    return [z[i] for z in l]
```

### Linear Simple :
- <font color='green'>Linear Model : OK</font>
- <font color='green'>MLP (2, 1) : OK</font>

```python
X = np.array([[1],[2]])
Y = np.array([2,3])

XF = X.flatten().tolist()
YF = Y.flatten().tolist()

inputCountPerSample = 1
mlpReg_1 = CLib.create_linear_model(inputCountPerSample)

mlpReg_1 = CLib.fit_regression(mlpReg_1, XF, YF, 1)

for x, y in zip(X, Y):
    plt.scatter(
        x,
        y,
        color='red'
    )

for i in range(0, 300):
    i /= 100
    value = CLib.predict_regression(mlpReg_1, [i]);
    plt.scatter(i, value, marker=',', color='blue', s=1)

plt.show()
plt.clf()
```

### Linear Simple 3D :
- <font color='green'>Linear Model : OK</font>
- <font color='green'>MLP (2, 1) : OK</font>

```python
X = np.array([ [1, 1], [2, 2], [3, 3] ]) 
Y = np.array([ 1, 2, 3 ])

XF = X.flatten().tolist()
YF = Y.flatten().tolist()

inputCountPerSample = 2
mlpReg_3 = CLib.create_linear_model(inputCountPerSample)

mlpReg_3 = CLib.fit_regression(mlpReg_3, XF, YF)

fig = plt.figure()
ax = Axes3D(fig)
ax.scatter(X[:,0], X[:,1], Y, color="red")

XX, YY, ZZ = [], [], []

for i in range(0, 1000, 10):
    for j in range(0, 1000, 10):
        i /= 100
        j /= 100
        XX.append(i)
        YY.append(j)
        zz = CLib.predict_regression(mlpReg_3, [i, j])
        ZZ.append(zz)

ax.plot_trisurf(XX, YY, ZZ, lw=0, color="grey", alpha=0.5)
plt.show()
```
